```python
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
x = np.arange(1, 51)
y = np.arange(10, 501, 10)
```

```python
y
```

```python
def gradient_descent(x,y):
    weight=0.5 #here i change the value of weight 
    bias=0.1
    lr=0.05
    n=len(x)
    itr=50
    
    for i in range(itr):
        # forward propogation
        y_hat = x * weight + bias
        mse = mean_squared_error(y, y_hat)
        
        #backward propogation
        dw = -(2/n) * sum(x * (y-y_hat))
        db = -(2/n) * sum(y-y_hat)
        
        weight = weight - (lr * dw)
        bias = bias - (lr * db)
        print(f"Itr ={i}, W={weight}, B={bias}, MSE={mse}")
```

```python
gradient_descent(x,y)
```
